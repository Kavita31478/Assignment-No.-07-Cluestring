# Hierachical Clustering

```python
# import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
crime = pd.read_csv("crime_data.csv")
crime
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(crime.iloc[:,1:])
df_norm
```

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='complete'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'complete')
```

```python
# save clusters for chart
y_hc = hc.fit_predict(df_norm)
y_hc
```

```python
crime['h_clusterid'] = y_hc
crime
```

# K-Mean Clustering

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import numpy as np
```

```python
crime1 = pd.read_csv("crime_data.csv")
crime1
```

```python
# Normalization function 
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)

#Normalized data frame (considering the numerical part of data)
df_norm=norm_func(crime1.iloc[:,1:])

df_norm.head(10)   #Top 10 rows
```

```python
# How to find optimum number of  cluster
#The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion:
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Selecting 4 clusters from the above pot which is the optimum number of custers
model=KMeans(n_clusters=4)
model.fit(df_norm)

model.labels_ #getting the labels of clusters assigned to each row
```

```python
md=pd.Series(model.labels_)#converting numpy array ntopandas series object
crime1['clust']=md #creating a new column and assigning it to new column
crime1
```

```python
#Murder -- Muder rates in different places of United States
#Assualt- Assualt rate in different places of United States
#UrbanPop - urban population in different places of United States
#Rape - Rape rate in different places of United States
crime1.iloc[:,1:6].groupby(crime1.clust).mean()
```

# DBSCAN

```python
#Import the libraries
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
# Import .csv file and convert it to a DataFrame object
df = pd.read_csv("crime_data.csv");

print(df.head())
```

```python
print(df.info())
```

```python
array=df.iloc[:,1:5]
```

```python
array.values
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
X
```

```python
dbscan = DBSCAN(eps=2, min_samples=5)
```

```python
dbscan.fit(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
```

```python
pd.concat([df,cl],axis=1)
```
